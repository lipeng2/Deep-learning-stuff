## transform date

```python
dataset['Year'] = list(map(lambda x: x.split('-')[0], dataset['Date'].squeeze().tolist()))
dataset['Month'] = list(map(lambda x:x.split('-')[1], dataset['Date'].squeeze().tolist()))
```

## rearrange column positions

```python
cols = dataset.columns.tolist()

new_cols = cols[-2:] + cols[:-2]

dataset = dataset[new_cols]
```

```python
dataset.head()
```

```python
dataset.columns = dataset.columns.get_level_values(0)
```

```python
dataset[['Open', 'High', 'Low', 'Close']] = dataset[['Open', 'High', 'Low', 'Close']].apply(pd.to_numeric) 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
dataset.groupby(['Year', 'Month']).mean().plot(figsize=(20,10))
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
x, y = [], []
tmp = df.Close.values
for i in range(10, len(tmp)):
    x.append(tmp[:i])
```

```python
def preproc_data(df, pred_var, his_sz=10, future_sz=1):
    print(f'Input length: {his_sz}, Output length: {future_sz}')
    df_needed = df[pred_var].values.reshape(-1,1)
    scaler = MinMaxScaler()
    df_needed = scaler.fit_transform(df_needed)
    x, y = [], []
    for i in range(len(df_needed)-his_sz-future_sz-1):
        x.append(df_needed[i:i+his_sz])
        y.append(df_needed[i+his_sz:i+his_sz+future_sz])
    return np.array(x), np.array(y), scaler
```

```python
x, y, scaler = preproc_data(dataset, 'Close')
```

```python
x.shape, y.shape
```

```python
x = x.reshape(-1, 1, 10)
y = y.reshape(-1, 1)
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# LSTM MODEL
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import LSTM

def build_model(his_sz, future_sz):
    model = Sequential()
    model.add(LSTM(32, input_shape=(1, his_sz), return_sequences = True))
    model.add(LSTM(16))
    model.add(Dense(future_sz, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer='adagrad')
    return model
```

```python
model = build_model(10,1)
```

```python
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss', patience=3)
```

```python
history = model.fit(X_train, y_train, epochs=30, batch_size=1, validation_split=0.2, shuffle=False, verbose=2, callbacks=[es])
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model train vs validation loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show()
```

```python
preds = scaler.inverse_transform(model.predict(x))
```

```python
plt.figure(figsize=(20,10))
plt.plot(scaler.inverse_transform(y)[::-1], label='actual')
plt.plot(preds[::-1], color='r', label='predictions')
plt.legend(loc='best')
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(preds, scaler.inverse_transform(y))
```

# 10 steps predictions

```python
step = 10
ps = 1
ep = 100
x1, y1, scaler = preproc_data(dataset, 'Close', step, ps)
```

```python
x1.shape, y1.shape
```

```python
x1 = x1.reshape(-1, 1, step)
y1 = y1.reshape(-1, ps)
```

```python
def build_model(his_sz, future_sz):
    model = Sequential()
    model.add(LSTM(32, input_shape=(1, his_sz), return_sequences = True))
    model.add(LSTM(16))
    model.add(Dense(32, activation='sigmoid'))
    model.add(Dense(future_sz, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer='adagrad')
    return model
```

```python
model2 = build_model(step, ps)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size=0.2)
```

```python
model2.fit(X_train, y_train, epochs=ep, batch_size=1, validation_split=0.2, shuffle=False, callbacks=[es], verbose=2)
```

```python
def plot_preds(x, y, i):
    plt.plot(model2.predict(x[i].reshape(-1,1,step)).squeeze(), label='pred')
    plt.plot(y[i], label='actual')
    plt.legend(loc='best')
```

```python
plot_preds(X_test, y_test, 2)
```

```python
preds = []
for i in range(0, len(X_test)-ps, ps):
    pred = model2.predict(X_test[i].reshape(1,1,step)).squeeze()
    preds.extend(pred)
```

```python
def trend_cal(vals):
    vals = list(vals)
    res = []
    for i in vals:
        res.append(1 if i[-1]-i[0]>0 else -1)
        
    return res
```
